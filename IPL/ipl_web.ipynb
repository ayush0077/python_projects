```python
def extract_data(data):
    cols=data.split('\n')[0].split()
    player=data.split('\n')[1:]
    player=np.array(player)
    no_of_player=int((player.size)/4)
    player=player.reshape((no_of_player,4))
    df=pd.DataFrame(player)
    other_info=df[3].str.split(expand=True)
    df=df.iloc[:,:-1]
    final=pd.concat([df,other_info],axis=1)
    cols=cols[:2]+['Temp']+cols[2:]
    final.columns=cols
    return final
```

```python
extract_data(data)
```

```python
driver=webdriver.Chrome()
import time 
all_data={}

for i in range(2008,2025):
    driver.get(f'https://www.iplt20.com/stats/{i}')
    time.sleep(10)
    data=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/div[4]/div/section/div/div[3]/div[2]/div[2]/div/div[3]/div/div[1]/table/tbody').text
    df=extract_data(data)
    df['Year']=i
    all_data[i]=df
    df.to_csv(f'E:\\Datascience\\dataset\\IPL\\{i}.csv')
    print(f'{i} year data extracted')
```
